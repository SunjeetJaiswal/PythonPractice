```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 12,6
import warnings
warnings.filterwarnings('ignore')
```

```python
vis3 = sns.lmplot(x='InternetUsers', y='BirthRate', data=stats )
```

```python
vis3 = sns.lmplot(x='InternetUsers', data=stats, y='BirthRate', fit_reg = False)
```

```python
vis3 = sns.lmplot(x='InternetUsers', data=stats, y='BirthRate', \
                  fit_reg = False, hue='IncomeGroup', size=10)
```

```python
vis3 = sns.lmplot(x='InternetUsers', data=stats, y='BirthRate', \
                  fit_reg = False, hue='IncomeGroup', size=10, aspect =1.44)
```
